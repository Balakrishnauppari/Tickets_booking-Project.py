```python
tickets = {
    "Kaleja": 50,
    "RRR": 40,
    "Bahubali": 30
} 

bookings = {} 

def show_tickets():
    print("\nAvailable Tickets:")
    for movie, count in tickets.items():
        print(movie,":" ,count)

def book_ticket():
    name = input("Enter your name: ")
    movie = input("Enter Movie: ") 
    if movie in tickets:
        if tickets[movie] > 0:
           tickets[movie] -= 1
           bookings[name] = movie 
           print("Tickets Booking Successfully ‚úÖ") 
        else:
            print("Sorry! Tickets are sold Out") 
    else:
        print("Movie not found ‚ùå")  

def cancel_ticket():
    name = input("Enter your name:") 

    if name in bookings:
        movie = bookings[name] 
        tickets[movie] += 1 
        del bookings[name] 
        print("Ticket cancelled successfully ‚úÖ")
    else:
        print("No booking found ‚ùå")    

def show_booking():
    if bookings:
        print("\nBooking Details:") 
        for name in bookings.items():
            print(name, "Booked", movie) 
    else:
        print("No booking yet") 

while True:
    print("\n--- Ticket Booking Menu ---")
    print("1. Show Tickets")
    print("2. Book Ticket")
    print("3. Cancel Ticket")
    print("4. Show Booking")
    print("5. Exit")        


    choice = input("Enter your Choice:") 

    if choice == "1":
        show_tickets() 
    elif choice == "2":
        book_ticket() 

    elif choice == "3":
        cancel_ticket()
    elif choice == "4":
        show_booking()
    elif choice == "5":
        print("Thank you for using Ticket Booking System üôè")
        break
    else:
        print("Invalid choice ‚ùå")    
    
```
